Preload linescans

```python
used_linescan_ids = [ linescan_datasets[id_ls].id for id_ls in matches_df.id_ls.unique() ]
linescans = { 
    linescan_id: dc.load(product='linescan', id=linescan_id, output_crs='epsg:28355', resolution=(-10,10)) 
        for linescan_id in used_linescan_ids
}
```

Merge linescans with associated polygons

```python
import os, contextlib

linescans_train = []
outputs_train = []

for id_ls, ls_df in matches_df.groupby(matches_df.id_ls):
    pl_ids = ls_df.id_pl.values
    
    linescan = linescans[ linescan_datasets[id_ls].id ]
    polygons_df = gdf[gdf.ID.isin(pl_ids)]
    
    with open(os.devnull, "w") as f, contextlib.redirect_stdout(f):
        tgt = xr_rasterize(gdf=polygons_df, da=linescan)

    linescans_train.append(linescan.linescan[0].data)
    outputs_train.append(tgt.data)
```

Dump linescans and polygons, resized to shaped used in training, to file

```python
resize256 = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((256, 256)),
])

stacked_linescans_train = torch.vstack([ resize256(ls) for ls in linescans_train ])
stacked_outputs_train = torch.vstack([ resize256(out) for out in outputs_train ])

torch.save(stacked_linescans_train, 'linescans_train.pt')
torch.save(stacked_outputs_train, 'outputs_train.pt')
```

## Training

```python
# https://github.com/mateuszbuda/brain-segmentation-pytorch/

class DiceLoss(torch.nn.Module):

    def __init__(self):
        super(DiceLoss, self).__init__()
        self.smooth = 1.0

    def forward(self, y_pred, y_true):
        assert y_pred.size() == y_true.size()
        y_pred = y_pred[:, 0].contiguous().view(-1)
        y_true = y_true[:, 0].contiguous().view(-1)
        intersection = (y_pred * y_true).sum()
        dsc = (2. * intersection + self.smooth) / (
            y_pred.sum() + y_true.sum() + self.smooth
        )
        return 1. - dsc
```

```python
from datetime import datetime
import torch
import shutil

# W oparciu o: medium.com/analytics-vidhya/saving-and-loading-your-model-to-resume-training-in-pytorch-cb687352fa61
def save_ckp(checkpoints_dir, epoch, model, optimizer, is_best = False):
    state = {
        'epoch': epoch,
        'state_dict': model.state_dict(),
        'optimizer': optimizer.state_dict()
    }
    
    f_path = checkpoints_dir / f'checkpoint_{datetime.now()}.pt'
    torch.save(state, f_path)
    if is_best:
        best_fpath = checkpoints_dir / 'best_model.pt'
        shutil.copyfile(f_path, best_fpath)

def load_ckp(checkpoint_fpath, model, optimizer):
    checkpoint = torch.load(checkpoint_fpath)
    model.load_state_dict(checkpoint['state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer'])
    return model, optimizer, checkpoint['epoch']

def load_model_from_ckp(checkpoint_fpath, model):
    checkpoint = torch.load(checkpoint_fpath)
    model.load_state_dict(checkpoint['state_dict'])
    return model, checkpoint['epoch']
```

```python
# W oparciu o: pytorch.org/tutorials/beginner/blitz/cifar10_tutorial.html
from pathlib import Path
import os    

CHECKPOINTS_DIR = 'chpts'
if not os.path.exists(CHECKPOINTS_DIR):
    os.makedirs(CHECKPOINTS_DIR)

import torch.optim as optim

model = LineUNet(usePretrained = False, in_channels = 1, init_features = 32)
# training_data = CustomDataset(transform = transforms.Resize((512, 512)), target_transform = transforms.Resize((512, 512)))
training_data = LinescansDataset()
trainloader = torch.utils.data.DataLoader(training_data, batch_size=32, shuffle=True)

def train(model, trainloader, device = 'cpu', niter = 10):
    criterion = DiceLoss() # torch.nn.MSELoss()
#     criterion = torch.nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)#, momentum=0.9)

    # transform = transforms.RandomApply(torch.nn.ModuleList([
    #     transforms.RandomAffine(degrees = 20, translate = (0.05, 0.05), scale = (0.95, 1.05), shear = 10)
    # ]), p = 0.5)
    augment_transform = NoneTransform()
    
    model.to(device)

    best_loss = np.inf
    model.train()

    for epoch in range(niter):  # loop over the dataset multiple times
        print(f"Epoch: {epoch}")

        running_loss = 0.0
        running_count = 0
        for i, data in enumerate(trainloader, 0):
            # get the inputs; data is a list of [inputs, labels]
            inputs, labels = data

            augmented = augment_transform(torch.vstack([inputs, labels]))
            inputs = augmented[:augmented.shape[0] // 2]
            labels = augmented[augmented.shape[0] // 2:]
            
            labels = labels

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward + backward + optimize
            outputs = model(inputs.to(device))
            loss = criterion(outputs, labels.to(device))
            loss.backward()
            optimizer.step()

            # print statistics
            running_loss += loss.item()
            running_count += inputs.shape[0]
    #     if i % 2000 == 1999:    # print every 2000 mini-batches
        print('[%d] loss: %.3f' %
              (epoch + 1, running_loss / running_count))

        if best_loss > running_loss or (epoch > 0 and epoch % 10 == 0):
            save_ckp(Path(CHECKPOINTS_DIR), epoch, model, optimizer, is_best = best_loss > running_loss)

        if best_loss > running_loss:
            best_loss = running_loss

    print('Finished Training')
```
